| Q1: Implement a pattern generator Input: a character (*, #, @) Input: shape type (triangle, pyramid, etc.) Input: height Output: printed ASCII shape Required OOP: base class + subclasses for each shape.

```python
def plot_triangle(height: int, left: bool = True):
    """
    Plot a triangle with * symbol.

    Args
    :height: Height of triangle
    :left: Pad by left side
    """
    for h in range(height):
        output = "*" * (h+1)
        if left:
            print(f"{output:<{height}}")    
        else:
            print(f"{output:>{height}}")

plot_triangle(5)
```

```python
def plot_pyramid(height: int):
    """
    Plot a pyramid with * symbol.

    If height = 3,

            *
           ***
          *****

    Args
    :height: Height of pyramid
    """
    for h in range(height):
        spaces = " " * (height - h - 1)
        print(spaces + "*" * (h * 2 + 1))

plot_pyramid(5)
```
